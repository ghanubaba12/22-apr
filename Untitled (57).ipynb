```python
Q1. Write a Python code to implement the KNN classifier algorithm on load_iris dataset in
sklearn.datasets.
ans-Sure! Here's an example Python code that implements the KNN classifier algorithm on the load_iris dataset from the sklearn.datasets module:

python
Copy code
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
iris = load_iris()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Initialize the KNN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Train the classifier on the training data
knn.fit(X_train, y_train)

# Predict the classes of the testing data
y_pred = knn.predict(X_test)

# Compute the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy:", accuracy)
In this code, we first load the iris dataset using the load_iris() function from sklearn.datasets. We then split the dataset into training and testing sets using the train_test_split() function from sklearn.model_selection.

Next, we initialize the KNN classifier with k=3 using the KNeighborsClassifier() function from sklearn.neighbors, and then train the classifier on the training data using the fit() method. We then use the trained classifier to predict the classes of the testing data using the predict() method.

Finally, we compute the accuracy of the classifier using the accuracy_score() function from sklearn.metrics and print the accuracy.





```

```python
Q3. Write a Python code snippet to find the optimal value of K for the KNN classifier algorithm using
cross-validation on load_iris dataset in sklearn.datasets.
ans-Certainly! Here's an example Python code that uses cross-validation to find the optimal value of K for the KNN classifier algorithm on the load_iris dataset from the sklearn.datasets module:

python
Copy code
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

# Load the dataset
iris = load_iris()

# Set the range of K values to test
k_range = range(1, 31)

# Initialize an empty list to store the cross-validation scores for each K value
k_scores = []

# Iterate through each K value and perform cross-validation
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, iris.data, iris.target, cv=10, scoring='accuracy')
    k_scores.append(scores.mean())

# Find the K value with the highest cross-validation score
optimal_k = k_range[k_scores.index(max(k_scores))]

# Print the optimal K value
print("Optimal K value:", optimal_k)
In this code, we first load the iris dataset using the load_iris() function from sklearn.datasets. We then set the range of K values to test using the range() function.

Next, we initialize an empty list k_scores to store the cross-validation scores for each K value. We then iterate through each K value and perform cross-validation using the cross_val_score() function from sklearn.model_selection. For each K value, we create a KNN classifier with that value of K and use the cross_val_score() function to compute the cross-validation scores for 10 folds of the data.

We then append the mean cross-validation score for each K value to the k_scores list. Once we have computed the cross-validation scores for all K values, we find the K value with the highest cross-validation score using the max() and index() functions.

Finally, we print the optimal K value. This value can be used to initialize a KNN classifier with the best K value for the load_iris dataset.





```

```python
Q4. Implement the KNN regressor algorithm with feature scaling on load_boston dataset in
sklearn.datasets.
ans-Sure, here's the Python code to implement the KNN regressor algorithm with feature scaling on the load_boston dataset in sklearn.datasets:

python
Copy code
# Import necessary libraries
from sklearn.datasets import load_boston
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Load the Boston Housing dataset
boston = load_boston()

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.2, random_state=42)

# Scale the features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a KNN regressor with k=5
knn = KNeighborsRegressor(n_neighbors=5)

# Fit the model on the scaled training data
knn.fit(X_train_scaled, y_train)

# Make predictions on the scaled test data
y_pred = knn.predict(X_test_scaled)

# Calculate the mean squared error of the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
In this code, we first load the Boston Housing dataset using load_boston() function from sklearn.datasets. We then split the dataset into train and test sets using the train_test_split() function from sklearn.model_selection.

Next, we scale the features using StandardScaler from sklearn.preprocessing. We fit the scaler on the training data using scaler.fit_transform(X_train) and transform both the training and test data using scaler.transform(X_test).

We then create a KNN regressor with n_neighbors=5 and fit the model on the scaled training data using knn.fit(X_train_scaled, y_train). We then make predictions on the scaled test data using knn.predict(X_test_scaled).

Finally, we calculate the mean squared error of the model using mean_squared_error(y_test, y_pred) from sklearn.metrics and print the result.





```

```python
Q5. Write a Python code snippet to implement the KNN classifier algorithm with weighted voting on
load_iris dataset in sklearn.datasets.
ans-Sure! Here's an example Python code that implements the KNN classifier algorithm with weighted voting on the load_iris dataset from the sklearn.datasets module:

python
Copy code
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
iris = load_iris()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)

# Initialize the KNN classifier with k=3 and use distance-based weights
knn = KNeighborsClassifier(n_neighbors=3, weights='distance')

# Train the classifier on the training data
knn.fit(X_train, y_train)

# Predict the classes of the testing data
y_pred = knn.predict(X_test)

# Compute the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy
print("Accuracy:", accuracy)
In this code, we first load the iris dataset using the load_iris() function from sklearn.datasets. We then split the dataset into training and testing sets using the train_test_split() function from sklearn.model_selection.

Next, we initialize the KNN classifier with k=3 and use distance-based weights by setting the weights parameter to 'distance'. This means that closer neighbors will have a greater influence on the classification decision.

We then train the classifier on the training data using the fit() method, and use the trained classifier to predict the classes of the testing data using the predict() method.

Finally, we compute the accuracy of the classifier using the accuracy_score() function from sklearn.metrics and print the accuracy. This code implements the KNN classifier algorithm with weighted voting on the load_iris dataset.





```

```python
Q7. Write a Python function to calculate the euclidean distance between two points.
ans-Sure! Here's an example Python function that calculates the Euclidean distance between two points:

python
Copy code
import math

def euclidean_distance(point1, point2):
    distance = 0.0
    for i in range(len(point1)):
        distance += (point1[i] - point2[i])**2
    return math.sqrt(distance)
This function takes two points as input and calculates the Euclidean distance between them. The points are represented as lists or tuples of numerical values.

The function first initializes a variable distance to 0.0. It then iterates through each dimension of the points using a for loop and calculates the squared difference between the corresponding values of the two points. It adds each squared difference to the distance variable.

Finally, the function returns the square root of the distance variable, which gives the Euclidean distance between the two points.

Here's an example of how to use this function:

python
Copy code
point1 = [1, 2, 3]
point2 = [4, 5, 6]
distance = euclidean_distance(point1, point2)
print(distance)
This would output 5.196152422706632, which is the Euclidean distance between the points [1, 2, 3] and [4, 5, 6].





```

```python
Q8. Write a Python function to calculate the manhattan distance between two points.
ans-ans-Sure! Here's a Python function to calculate the Manhattan distance between two points:

python
Copy code
def manhattan_distance(point1, point2):
    """
    Calculates the Manhattan distance between two points.
    
    Parameters:
    -----------
    point1 : tuple or array-like
        The coordinates of point 1
    point2 : tuple or array-like
        The coordinates of point 2
        
    Returns:
    --------
    distance : float
        The Manhattan distance between point1 and point2
    """
    # Calculate the absolute differences between the coordinates of the two points
    diff = abs(point1 - point2)
    
    # Calculate the Manhattan distance as the sum of the absolute differences
    distance = diff.sum()
    
    return distance
This function takes two points point1 and point2 as input, where each point can be represented as a tuple or array-like object with the same number of dimensions. It then calculates the absolute differences between the coordinates of the two points using abs(point1 - point2) and sums them up to calculate the Manhattan distance between the two points.

You can use this function to calculate the Manhattan distance between two points in your KNN algorithm implementation.





```
